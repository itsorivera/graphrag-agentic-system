### LiteLLM AWS Bedrock

```python
from adapter.repository.llm_provider.LiteLLMProviderAdapter import LiteLLMProviderAdapter
from config.app_config import AppConfig
from rich import print as rprint

app_config = AppConfig()

llm_adapter = LiteLLMProviderAdapter()
llm = llm_adapter.get_bedrock_llm(app_config.DEFAULT_MODEL_ID)
```

```python
from google.adk.models.llm_request import LlmRequest
from google.genai import types

request = LlmRequest(
    contents=[
        types.Content(
            role="user",
            parts=[types.Part.from_text(text='What is the meaning of life?')]
        )
    ]
)
full_text = ""
async for response in llm.generate_content_async(request):
    if response.content and response.content.parts:
        for part in response.content.parts:
            if part.text:
                chunk_text = part.text
                #print(chunk_text, end="", flush=True)
                full_text += chunk_text

rprint(f"\n\nResponse: {full_text}")
```
